# Types of Graphs
1)bar plot
2)sctter plot
3)pie plot
4)histogram

```python
a=np.array(['a','b','c'])
b=np.array([1,3,2])
plt.bar(a,b)
```

```python
p=np.array([1,2,3,4,5])
plt.hist(p)
```

```python
a=np.arange(1,10,2)
b=np.arange(0,10,2)
plt.bar(a,[1,2,3,4,5])
plt.bar(b,[6,7,8,9,10])
```

```python
plt.plot(a)
plt.plot(b)
plt.plot(p)
```

```python
x=np.array([1,1.5,2,2.5,3,3.5,3.6])
y=np.array([7.5,8,8.5,9,9.5,10,10.5])


x1=np.array([4,4.5,5,5.5,6,6.6,10])
y1=np.array([1,2,3,4.4,5.5,6.6,7.7])
plt.scatter(x,y,label='label1')
plt.scatter(x1,y1,label='label2')
plt.legend()
```

```python
lab=np.array(['a','b','c','d','e','f','g'])
exp=[0.3,0,0.2,0,0,0,0]
plt.pie(x,labels=lab,explode=exp)
plt.legend()
```
