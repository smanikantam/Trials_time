```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv('cricket.csv')
data
```

```python
def cov_num(col):
  i=1
  for value in data[col].unique():
    data[col]=data[col].replace(value,str(i))
    i+=1
  data[col]=data[col].astype(float)
  data[col].unique()
```

```python
data.isnull().any()
```

```python
data=data.ffill()
# ffill
# bfill
# fillna
```

```python
data.isnull().any()
```

```python
data['Result'].unique()
```

```python
data['Result']=data['Result'].replace('Lose',str(1))
data['Result']=data['Result'].replace(' Lose D/L',str(1))
data['Result']=data['Result'].replace('Win',str(2))
data['Result']=data['Result'].replace('win',str(2))
data['Result']=data['Result'].replace(' Win D/L',str(2))
data['Result']=data['Result'].replace('No Result',str(0))
```

```python
data['Result']=data['Result'].astype(float)
```

```python
data['Toss']=data['Toss'].replace('win','Win')
```
